```python
import pandas as pd
import plotly.express as px
```

```python
asset_data = pd.read_csv("EF_Data_Summary.csv")
asset_data
```

```python
asset_data_core = asset_data
asset_data_core
```

```python
#Add "monthly return" to all columns besides Month
asset_data.columns = [col + " Monthly Return" if col != "Month" else col for col in asset_data.columns]

# Make every value a pct change except for the 'Month' column
asset_data.loc[:, asset_data.columns != "Month"] = asset_data.loc[:, asset_data.columns != "Month"].pct_change()

asset_data_returns = asset_data.loc[:, asset_data.columns.str.contains("Return")]
asset_data_returns = asset_data_returns.dropna()
asset_data_returns
```

```python
asset_data_returns.tail(12)
```

```python
#I added this case for one year (11 months) but I am still confused. Ask Mark to help
def expected_return(rows):
    returns = asset_data_returns.tail(rows).mean()
    if rows!=11:
        returns_annualized = (1 + returns)**12 - 1
    else:
        returns_annualized = returns * 12
    return returns_annualized
```

```python
def st_dev(rows):
    vol = asset_data_returns.tail(rows).std()
    vol_annualized = vol * 12**0.5
    return vol_annualized
```

```python
returns_annualized_10 = expected_return(119)
returns_annualized_10
```

```python
vol_annualized_10 = st_dev(119)
vol_annualized_10
```

```python
assets = pd.read_csv("EF_Data_Summary.csv")
```

```python
def historical_graph(asset_df):
    fig = px.line(asset_df, x="Month", y=["S&P 500", "Fixed Income", "Gold", "Private Credit", "Real Estate", "Private Equity"], 
              title="Historical Asset Prices",
              labels={"value": "Price", "variable": "Asset"},
              markers=True)
    return fig
```
